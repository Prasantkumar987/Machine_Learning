```python
fea = ['battery_power', 'dual_sim', 'four_g',
       'int_memory',  'mobile_wt', 'n_cores', 'ram', 
       'touch_screen']
X = data[fea]
y = data['price_range']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline

from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import ExtraTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
```

```python
etc_pl = Pipeline([('scaling0',StandardScaler()),
                  ('etc',ExtraTreesClassifier())])
```

```python
et_pl = Pipeline([('scaling1',StandardScaler()),
                  ('et',ExtraTreeClassifier())])
```

```python
nb_pl = Pipeline([('scaling2',StandardScaler()),
                  ('nb',GaussianNB())])
```

```python
svm_pl = Pipeline([('scaling3',StandardScaler()),
                   ('svm',SVC(kernel='linear',random_state=0))])
```

```python
knn_pl = Pipeline([('scaling4',StandardScaler()),
                   ('knn',KNeighborsClassifier(n_neighbors=5,metric = 'minkowski',p=2))])
```

```python
rf_pl = Pipeline([('scaling5',StandardScaler()),
                   ('rf',RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0))])
```

```python
bg_pl = Pipeline([('scaling6',StandardScaler()),
                  ('bg',BaggingClassifier())])
```

```python
gc_pl = Pipeline([('scaling7',StandardScaler()),
                  ('gc',GaussianProcessClassifier())])
```

```python
pipelines = [etc_pl,et_pl,nb_pl,svm_pl,knn_pl,rf_pl,bg_pl,gc_pl]
```

```python
best_accuracy = 0.0
best_pipeline = ""
```

```python
for pipe in pipelines:
  pipe.fit(X_train,y_train)
```

```python
for pipe in pipelines:
  if pipe.score(X_test,y_test)>best_accuracy:
    best_accuracy = pipe.score(X_test,y_test)
    best_pipeline = pipe
print("Classifier with Best accuracy : {}".format(best_pipeline))
```
